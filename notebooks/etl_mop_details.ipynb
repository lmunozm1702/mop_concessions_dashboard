#### Column "region"

```python
def get_region_number(x):
    for i in x:
        if i.find("arica") != -1:
            x[x.index(i)] = str(15)
        elif i.find("tarapaca") != -1:
            x[x.index(i)] = str(1)
        elif i.find("antofagasta") != -1:
            x[x.index(i)] = str(2)
        elif i.find("atacama") != -1:
            x[x.index(i)] = str(3)
        elif i.find("coquimbo") != -1:
            x[x.index(i)] = str(4)
        elif i.find("valparaiso") != -1:
            x[x.index(i)] = str(5)
        elif i.find("metropolitana") != -1:
            x[x.index(i)] = str(13)
        elif i.find("o'higgins") != -1:
            x[x.index(i)] = str(6)
        elif i.find("maule") != -1:
            x[x.index(i)] = str(7)
        elif i.find("biobio") != -1 or i.find("bio bio") != -1: 
            x[x.index(i)] = str(8)
        elif i.find("araucania") != -1:
            x[x.index(i)] = str(9)
        elif i.find("lagos") != -1:
            x[x.index(i)] = str(10)
        elif i.find("rios") != -1:
            x[x.index(i)] = str(14)
        elif i.find("aysen") != -1:
            x[x.index(i)] = str(11)
        elif i.find("magallanes") != -1:
            x[x.index(i)] = str(12)
        elif i.find("nuble") != -1:
            x[x.index(i)] = str(16)
        elif i.find("territorio nacional") != -1:
            x[x.index(i)] = str(17)
    return x


def extract_region_data(x):
    x = x.lower()
    x = unidecode(x)
    x = x.replace("arica y parinacota", "arica_y_parinacota")
    x = x.replace("es y an", "es_y_an")
    print(x)
    x = re.split(",| y ", x)
    x = [i.strip() for i in x]
    x = get_region_number(x)
    x.sort(key=int)
    return ",".join(x)
```

```python
df["region"] = df["region"].apply(lambda x: extract_region_data(x))
df["region"].value_counts()
```

#### Column "provincia"

```python
def extract_provincia_data(x):
    x = x.lower()
    x = unidecode(x)
    print(x)
    x = re.split(",| y ", x)
    x = [i.strip() for i in x]
    x.sort()
    return ",".join(x)
```

```python
df["provincia"].value_counts()
df["provincia"] = df["provincia"].apply(lambda x: extract_provincia_data(x))
df["provincia"].value_counts()
```

#### Column "comuna"

#### Column "presupuesto oficial", we will create 2 new columns to extract the "UF" and "USD" values

```python
def get_uf(x):
    x = re.findall(r"(UF )([0-9\.]+)(.*)", x)
    if len(x) > 0:
        return x[0][1].replace(".", "").strip()
    else:
        return 0
    

def get_usd(x):
    x = re.findall(r"(.*)(USD\D+)([0-9\.\,]+)(.*)", x)
    if len(x) > 0:
        return x[0][2].replace(".", "").replace(",", ".").strip()
    else:
        return 0

df["UF"] = df["presupuesto oficial"].apply(lambda x: get_uf(x))
df["UF"] = df["UF"].astype(float)
print(df["UF"])


df["USD"] = df["presupuesto oficial"].apply(lambda x: get_usd(x))
df["USD"] = df["USD"].astype(float)
print(df["USD"])
```
